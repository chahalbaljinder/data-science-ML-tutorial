K NEAREST NEIGHBOUR ALGO

```python
import pandas as pd
import matplotlib.pyplot as plt 
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv("Social_Network_ads.csv")
dataset.head(3)
```

```python
dataset.drop(columns=["User ID","Gender"],inplace=True)
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Purchased"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
x=pd.DataFrame(ss.fit_transform(x),columns=x.columns)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn=KNeighborsClassifier(n_neighbors=4) #by default p=2 (1= manhattan dist. and 2=euclideans dist.)
kn.fit(x_train,y_train)
```

```python
for i in range (1,30):
    kn1=KNeighborsClassifier(n_neighbors=i)   #for finding best fit for data
    kn1.fit(x_train,y_train)
    print (i , kn1.score(x_train,y_train)*100 , kn1.score(x_test,y_test)*100)  
```

```python
kn.score(x_test,y_test)*100 , kn.score(x_train,y_train)*100  #regular method of finding best fit by hit and trail
```

```python
kn.predict([[-1.781797 ,	-1.490046]])
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=kn)
```

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
dataset=pd.read_csv("Salary_data.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
x=dataset.drop(columns=["Salary"])
y=dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knr=KNeighborsRegressor(n_neighbors=3)
knr.fit(x_train,y_train)
```

```python
knr.score(x_test,y_test)*100
```

```python
knr.score(x_train,y_train)*100
```

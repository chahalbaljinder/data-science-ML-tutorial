STOP WORD REMOVAL

```python
x = "Natural language processing (NLP) is an interdisciplinary subfield of computer science and artificial intelligence. It is primarily concerned with providing computers the ability to process data encoded in natural language and is thus closely related to information retrieval, knowledge representation and computational linguistics, a subfield of linguistics. Typically data is collected in text corpora, using either rule-based, statistical or neural-based approaches of machine learning and deep learning."
```

```python
from nltk.tokenize import word_tokenize
```

```python
word = word_tokenize(x)
new_word = list(word)
new_word
```

```python
from nltk.corpus import stopwords
```

```python
stop = stopwords.words("english")
stop
```

```python
from string import punctuation
```

```python
stop_words = list (punctuation) + stop
stop_words
```

```python
for i in word:
    if i not in stop_words:
        print (i)
```

STEMMING 

```python
from nltk.stem import LancasterStemmer , SnowballStemmer , PorterStemmer , RegexpStemmer 
```

```python
l = LancasterStemmer()
lan_stemmer=[l.stem(word) for word in new_word]
lan_stemmer
```

```python
s = SnowballStemmer("english")
sb_stemmer = [s.stem(word) for word in new_word]
sb_stemmer
```

```python
p = PorterStemmer()
porter_stemmer = [p.stem(word) for word in new_word]
porter_stemmer
```

```python
r = RegexpStemmer("ing|ion" , min=5)   #min is for the minimum length of string to stem
reg_stemmer = [r.stem(word) for word in new_word]
reg_stemmer
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
new_word1 = ["playing", "running", "explanation", "station", "catting"]
```

```python
wl = WordNetLemmatizer()
wnl=[wl.lemmatize(word , pos = "v") for word in new_word1]   #pos define which part of speech should be used on the words
wnl
```

N-GRAMS 

```python
from nltk.util import ngrams
from nltk.collocations import ngrams , BigramCollocationFinder , TrigramCollocationFinder
```

```python
n = ngrams(new_word , 4) #by default 1
for i in n:
    print(i)
```

```python
b = BigramCollocationFinder.from_words(new_word) #occurancies of pairs of 2 element
b.ngram_fd
```

```python
t = TrigramCollocationFinder.from_words(new_word)  #occurancies of pairs of 3 element
t.ngram_fd
```

COUNT VECTORIZER

```python
doc = ["The cat sat on the mat.",
    "The dog ate my homework.",
    "The cat chased the dog."]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(input="content")  #for inserting content,  input: {'filename', 'file', 'content'}, default='content'
x = cv.fit_transform(doc)

x_arr = x.toarray()

feature_names = cv.get_feature_names_out() # for getting features name in alphabetical order

print(feature_names)
print(x_arr)
```

```python
cv.vocabulary_  #indexing on the basis of alphabetical order
```

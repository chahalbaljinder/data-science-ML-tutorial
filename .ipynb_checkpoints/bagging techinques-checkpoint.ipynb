CLASSIFIER 

```python
from sklearn.datasets import make_moons
import pandas as pd 
import seaborn as sns 
```

```python
x,y=make_moons(n_samples=1000 , noise=0.5)
```

```python
df = {"x1":x[:,0] , "x2":x[:,1]  , "y":y}
dataset=pd.DataFrame(df)
dataset.head(3)
```

```python
sns.scatterplot(x="x1" , y="x2" , data=dataset , hue="y")
```

```python
x=dataset.iloc[:,:-1]
y=dataset["y"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

BAGGING META ESTIMATOR

```python
from sklearn.ensemble import BaggingClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
sv=SVC()
sv.fit(x_train,y_train)
```

```python
sv.score(x_train,y_train)*100  ,  sv.score(x_test,y_test)*100 
```

```python
bc = BaggingClassifier(estimator=SVC(),n_estimators=10)
bc.fit(x_train,y_train)
```

```python
bc.score(x_train,y_train)*100  ,  bc.score(x_test,y_test)*100 
#there is slight change in accuracy 
```

RANDOM FOREST

```python
rf=RandomForestClassifier(n_estimators=100)  #will only take no. of estimators not the estimator coz it has by default estimator as decision tree
rf.fit(x_train, y_train)
```

```python
rf.score(x_train,y_train)*100 , rf.score(x_test,y_test)*100
```

```python
import pandas as pd
```

```python
dataset=pd.read_csv("Placement.csv")
dataset.head(3)
```

```python
x=dataset[["cgpa"]]
y=dataset["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
svr=SVR()
svr.fit(x_train,y_train)
```

```python
svr.score(x_train,y_train)*100 , svr.score(x_test,y_test)*100
```

```python
br=BaggingRegressor(estimator=SVR(),n_estimators=10)
br.fit(x_train,y_train)
```

```python
br.score(x_train,y_train)*100 , br.score(x_test,y_test)*100
```

```python
rfr=RandomForestRegressor(n_estimators=50)
rfr.fit(x_train,y_train)
```

```python
rfr.score(x_train,y_train)*100 , rfr.score(x_test,y_test)*100
```

KMEANS CLUSTERING

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
dataset=pd.read_csv("iris_raw.csv")
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(3,3))
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]

for i in range (2,21):
    km=KMeans(n_clusters=i, init="k-means++")
    km.fit(dataset)
    wcss.append(km.inertia_) #sum of squared distances from each point to its assigned cluster center,
                             #which is also known as the within-cluster sum of squares (WCSS)
wcss
```

```python
plt.figure(figsize=(10,7))
plt.plot([i for i in range (2,21)],wcss,marker="o")  #marker is for marking each point
plt.xlabel("no. of clusters")  
plt.xticks([i for i in range (2,21)])  #for marking axis points 
plt.grid(axis="x")  #for making grid 
plt.grid(axis="y")
plt.ylabel("wcss")
plt.show()
#choose the point from where the elbow starts bending 
```

```python
kmn = KMeans(n_clusters=3) #no. of clusters taken from graph
dataset["Predict"]=kmn.fit_predict(dataset)  #for predicting the cluster elements
dataset.head(3)
```

```python
kmn.labels_
```

SILHOUETTE SCORE DIRECT APPORACH 

If silhouette_score is towards 1 then it is best clustering and if towards -1 then it is bad clustering 

```python
from sklearn.metrics import silhouette_score
silhouette_score(dataset,labels=kmn.labels_)
```

Loop and through graph checking for how many clusters should be made

```python
ss=[]
no_c=[j for j in range (2,21)]

for i in range (2,21):
    km1=KMeans(n_clusters=i)
    km1.fit(dataset)
    ss.append(silhouette_score(dataset,km1.labels_))
```

```python
plt.plot(no_c , ss)
plt.xlabel("no. of clusters")
plt.ylabel("sihouette_score")
plt.xticks(no_c)
plt.grid(axis="x")
plt.show()
#corresponding to the highest point is no. of clusters
```

SILHOUETTE SCORE ENDED HERE 

```python
sns.pairplot(data=dataset,hue="Predict")
plt.savefig("prd_iris")
plt.show()
```

HIERARCHICAL CLUSTERING 
THROUGH AGGLOMERATIVE CLUSTERING 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
dataset=pd.read_csv("iris_raw.csv")
```

```python
dataset.head(3)
```

```python
sns.pairplot(data=dataset) #check for linearly separable data
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,20))
sc.dendrogram(sc.linkage(dataset,method='single',metric='euclidean'))  #method and metric can be changed for betterment
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac=AgglomerativeClustering(n_clusters=2 , linkage="single")
dataset["predict"]=ac.fit_predict(dataset)
dataset.head(3)
```

```python
sns.pairplot(data=dataset,hue="predict")
plt.show()
```

DBSCAN CLUSTERING

```python
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.datasets import make_moons
import pandas as pd 
```

```python
x, y = make_moons(n_samples=250, noise=0.05)   #dataset creation for DBSCAN 
```

```python
dataset=pd.DataFrame({"data1":x[:,0] , "data2":x[:,1]})
dataset.head(2)
```

```python
sns.scatterplot(x="data1", y="data2", data=dataset)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.2 , min_samples=4)  #change params for better clustering 
dataset["predict"]=db.fit_predict(dataset)
```

```python
sns.scatterplot(x="data1", y="data2", data=dataset,hue="predict")
plt.show()
```

```python
sns.scatterplot(x="data1", y="data2", data=dataset,hue=y) #y is from dataset
plt.show()                                                #match both original and predicted data graphs
```

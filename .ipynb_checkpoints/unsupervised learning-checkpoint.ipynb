```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
dataset=pd.read_csv("iris_raw.csv")
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(3,3))
sns.pairplot(data=dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]

for i in range (2,21):
    km=KMeans(n_clusters=i, init="k-means++")
    km.fit(dataset)
    wcss.append(km.inertia_) ''sum of squared distances from each point to its assigned cluster center,
                             which is also known as the within-cluster sum of squares (WCSS)'''
wcss
```

```python
plt.figure(figsize=(10,7))
plt.plot([i for i in range (2,21)],wcss,marker="o")  #marker is for marking each point
plt.xlabel("no. of clusters")  
plt.xticks([i for i in range (2,21)])  #for marking axis points 
plt.grid(axis="x")  #for making grid 
plt.grid(axis="y")
plt.ylabel("wcss")
plt.show()
#choose the point from where the elbow starts bending 
```

```python
kmn = KMeans(n_clusters=3) #no. of clusters taken from graph
dataset["Predict"]=kmn.fit_predict(dataset)  #for predicting the cluster elements
dataset.head(3)
```

```python
sns.pairplot(data=dataset,hue="Predict")
plt.savefig("prd_iris")
plt.show()
```

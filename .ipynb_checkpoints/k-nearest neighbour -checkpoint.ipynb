K NEAREST NEIGHBOUR ALGO

```python
import pandas as pd
import matplotlib.pyplot as plt 
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv("Social_Network_ads.csv")
dataset.head(3)
```

```python
dataset.drop(columns=["User ID","Gender"],inplace=True)
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Purchased"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
x=pd.DataFrame(ss.fit_transform(x),columns=x.columns)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn=KNeighborsClassifier(n_neighbors=4) #by default p=2 (1= manhattan dist. and 2=euclideans dist.)
kn.fit(x_train,y_train)
```

```python
for i in range (1,30):
    kn1=KNeighborsClassifier(n_neighbors=i)   #for finding best fit for data
    kn1.fit(x_train,y_train)
    print (i , kn1.score(x_train,y_train)*100 , kn1.score(x_test,y_test)*100)  
```

```python
kn.score(x_test,y_test)*100 , kn.score(x_train,y_train)*100  #regular method of finding best fit by hit and trail
```

```python
kn.predict([[-1.781797 ,	-1.490046]])
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=kn)
```

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Placement.csv")
dataset.head(3)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="cgpa",y="package",data=dataset)
```

```python
x=dataset[["cgpa"]]
y=dataset["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVR
```

```python
#types of kernels -- linear , poly , rbf , sigmoid , precomputed || precomputed works on sparse matrix
```

```python
sr=SVR(kernel="linear")
sr.fit(x_train,y_train)
```

```python
sr.score(x_test,y_test)*100
```

```python
sr.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="cgpa",y="package",data=dataset)
plt.plot(dataset[["cgpa"]],sr.predict(x),c="red")
```

HYPERPARAMETER TUNING

```python
import pandas as pd
```

```python
dataset=pd.read_csv("Salary_data.csv")
dataset.head(3)
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor(criterion= 'poisson', max_depth = 13, splitter = 'random')
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100 , dt.score(x_train,y_train)*100
```

```python
from sklearn.model_selection import GridSearchCV 
```

```python
df={"criterion":['squared_error','friedman_mse', 'absolute_error', 'poisson'], 
    "splitter":['best', 'random'], 
    "max_depth":[i for i in range (2,20)]}
```

```python
gd=GridSearchCV(DecisionTreeRegressor(),param_grid=df)
gd.fit(x_train,y_train)
```

```python
gd.best_params_  #will give 4*2*18 (criterion * splitter * max_depth (18)) which means 144 models  and will given different output each time form 144 models 
```

```python
gd.best_score_
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor(criterion= 'absolute_error', max_depth = 13, splitter = 'random')
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100 , dt.score(x_train,y_train)*100
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
df={"criterion":['squared_error','friedman_mse', 'absolute_error', 'poisson'], 
    "splitter":['best', 'random'], 
    "max_depth":[i for i in range (2,20)]}
```

```python
rs=RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=df , n_iter=20)  #n_iter means will give 20 random samples 
rs.fit(x_train,y_train)
```

```python
rs.best_params_ #will change each time on running but will repeat between random 2 samples 
```

```python
rs.best_score_
```

APRIORI ALGO. MARKET BASKET ANALYSIS

```python
import pandas as pd 
```

```python
dataset=pd.read_csv("groceries - groceries.csv")
```

```python
dataset.head(3)
```

```python
dataset.shape
```

Removing nan values for dataset cleaning
nan values are of float datatype

```python
# Initialize an empty list called 'market' which will store the processed data.
market = []
# Iterate over each row in the dataset using its index.
for i in range(0,dataset.shape[0]):
    cus=[]
    # Iterate over each column in the dataset.
    for j in dataset.columns:
        # Check if the value in the current cell is of string type.
        if type(dataset[j][i])== str:
            # Append the string value to the 'cus' list.
            cus.append(dataset[j][i])
    market.append(cus)
```

```python
market
```

```python
l=[]
for i in market:
    for j in i:           #above list is in list in list form so from this we can make it a single list
        l.append(j)
l
```

```python
import collections 
```

```python
p=collections.Counter(l) #return the occurany of each item
```

```python
p.keys()
```

```python
p.values()
```

```python
d={"Item_name": p.keys() , "values" : p.values()}    #creating dictionary for dataframe 
pd.DataFrame(d)                      #Dataframe creation
pd.set_option("display.max_rows",200)  #for displaying all rows 
```

```python
pd.DataFrame(d).sort_values(by=["values"],ascending=False)  #sorting in descending order on basis of values 
#checking for most selling product
```

APRIORI STARTS TILL NOW WE WERE CLEANING OUR DATA

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
te = TransactionEncoder()
te.fit(market)
```

```python
te.transform(market)
```

```python
df = pd.DataFrame(te.transform(market) , columns=te.columns_)  #making dataframe to apply apriori 
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(df, min_support=0.05, use_colnames=True, max_len=None).sort_values(by=["support"])
#from here we can find the best pattern for placing items 
```

F-P GROWTH AGORITHM (FREQUENT PATTERN)

```python
#ALL STEPS ARE SAME AS APRIORI ALGORITHM TILL TRANSENCODING 
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

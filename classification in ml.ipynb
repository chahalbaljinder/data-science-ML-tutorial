IMBALANCED DATASET HANDLING

```python
import pandas as pd 
```

```python
dataset=pd.read_csv("suv_data.csv")
```

```python
dataset.head(3)
```

```python
dataset["Purchased"].value_counts()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[32,120000]])  #from this we can conclude that out model is biased due to imbalanced dataset
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
ru=RandomUnderSampler()
ru_x,ru_y=ru.fit_resample(x,y)
```

```python
ru_y.value_counts()  #for checking balanced dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(ru_x,ru_y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[48 , 29000]])
```

```python
dataset=pd.read_csv("suv_data.csv")
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Purchased"]
y.value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro=RandomOverSampler()
ro_x,ro_y=ro.fit_resample(x,y)
```

```python
ro_y.value_counts()  #the data is balanced now
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(ro_x,ro_y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[47,49000]])
```

NAIVE BAYES 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
from mlxtend.plotting import plot_decision_regions
import seaborn as sns 
```

```python
dataset=pd.read_csv("Placement1.csv")
dataset.head(3)
```

```python
dataset.drop(columns=["Student_ID"],inplace=True)
dataset.head(3)
```

```python
plt.subplot(1,2,1)
sns.kdeplot(dataset["CGPA"])
                                    #for checking nature of the data
plt.subplot(1,2,2)                   
sns.kdeplot(dataset["IQ"])
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB , MultinomialNB , BernoulliNB
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100  ,  gnb.score(x_train,y_train)*100
```

```python
gnb.predict()
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
```

```python
mnb = MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100 , mnb.score(x_train,y_train)*100
```

```python
mnb.predict()
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
```

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100 , bnb.score(x_train,y_train)*100
```

```python
bnb.predict()
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
```

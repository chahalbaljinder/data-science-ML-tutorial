```python
from sklearn.preprocessing import FunctionTransformer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("loan_prediction.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
q1=dataset["CoapplicantIncome"].quantile(0.25)
q3=dataset["CoapplicantIncome"].quantile(0.75)
iqr=q3-q1
iqr,q1,q3
```

```python
min_r=q1-(1.5*iqr)
max_r=q3+(1.5*iqr)
min_r,max_r
```

```python
dataset=dataset[dataset["CoapplicantIncome"]<=max_r]
```

```python
plt.figure(figsize=(4,3))
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
#Now we check the graph after function transformation\
ft=FunctionTransformer(func=np.log1p)   #function can be any function according to the data
dataset["CoapplicantIncome_ft"]=ft.fit_transform(dataset["CoapplicantIncome"])
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("after")
sns.distplot(dataset["CoapplicantIncome_ft"])

plt.show()

```

```python
import pandas as pd 
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
dataset=pd.read_csv("Diabetes.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
x=dataset.iloc[:,:-1] #used for integer-location based indexing for selection by position. [:,:-1]will select all columns except the last one. 
y=dataset["Outcome"]  
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
fs=SequentialFeatureSelector(lr,k_features=6,forward=True)  #k_features must be (all columns-1)
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_   #by changing k_feature in fs check for the highest accuracy
```

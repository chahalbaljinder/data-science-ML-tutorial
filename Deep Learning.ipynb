```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Placement1.csv")
```

```python
dataset.drop(columns=["Student_ID"],inplace=True)
dataset.head(3)
```

```python
sns.scatterplot(x="CGPA" , y="IQ"  , data=dataset , hue="Placement")
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=70)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron(alpha=0.001)
pr.fit(x_train , y_train)
```

```python
pr.score(x_train , y_train)*100  ,  pr.score(x_test , y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf=pr)
```
